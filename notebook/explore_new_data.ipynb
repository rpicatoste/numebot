# Explore data v4.1

```python
from pathlib import Path

current_round = napi.get_current_round()
data_path = Path('/home/pica/nas_pica/Data/numerai/data/v4.1/') / str(current_round)

from dataclasses import dataclass

@dataclass
class DataFileNames:
    validation: str = 'validation.parquet'
    train: str = 'train.parquet'
    live: str = 'live.parquet'
    live_example_preds: str = 'live_example_preds.parquet'
    validation_example_preds: str = 'validation_example_preds.parquet'
    features: str = 'features.json'
    meta_model: str = 'meta_model.parquet'
    
fn = DataFileNames()
```

```python
napi.download_dataset(filename=f"v4.1/{fn.validation}", dest_path=str(data_path/fn.validation))
napi.download_dataset(filename=f"v4.1/{fn.train}", dest_path=str(data_path/fn.train))
napi.download_dataset(filename=f"v4.1/{fn.live}", dest_path=str(data_path/fn.live))
napi.download_dataset(filename=f"v4.1/{fn.live_example_preds}", dest_path=str(data_path/fn.live_example_preds))
napi.download_dataset(filename=f"v4.1/{fn.validation_example_preds}", dest_path=str(data_path/fn.validation_example_preds))
napi.download_dataset(filename=f"v4.1/{fn.features}", dest_path=str(data_path/fn.features))
napi.download_dataset(filename=f"v4.1/{fn.meta_model}", dest_path=str(data_path/fn.meta_model))
```
